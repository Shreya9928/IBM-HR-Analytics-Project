```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('abc.csv')
```

```python
df.info()
df.describe()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop(['EmployeeCount', 'Over18','EmployeeNumber','StandardHours'], axis= 1, inplace=True)
```

```python
sns.countplot(x="Attrition",data=df,palette="coolwarm",)
plt.title("Attrition count plot")
plt.show()
```

```python
sns.countplot(x="Gender",hue="Attrition",data=df,palette="Set1",)
plt.title("attrition count by gender")
plt.show()

```

```python
sns.countplot(x="Department",hue="Attrition",data=df,palette="Set2",)
plt.title("Attrition count by gender")
plt.show()
```

```python
sns.boxplot(x="JobRole",y="MonthlyIncome",data=df,palette="Set2",hue="Attrition")
plt.xticks(rotation=45)
plt.title("Income Distribution by Job Role and Attrition")
plt.show()
```

```python
sns.countplot(x="JobSatisfaction",hue="Attrition",data=df,palette="Set3",)
plt.title("Attrition Count by gender")
plt.show()
```

```python
corr = df.select_dtypes(include='number').corr()
plt.figure(figsize=(15,10))
sns.heatmap(corr, annot=True, cmap='coolwarm',fmt='.2f')
plt.title("Feature Correlation Heatmap")
```

```python
sns.boxplot(x="JobLevel",y="MonthlyIncome",data=df)
plt.title("Monthly Income by Job level")
plt.show()
```

```python
sns.boxplot(x="YearsAtCompany",y="YearsInCurrentRole",data=df)
plt.show()
```

```python
bins=[18,25,35,45,60]
labels=['18-25','25-35','36-45','46-60']
df["age_group"]=pd.cut(df['Age'],bins=bins,labels=labels)
```

```python
sns.countplot(x="age_group",data=df,hue="Attrition",palette="magma")
plt.title("Attrition by Age Group")
plt.show()
```

```python
sns.countplot(data=df, x='WorkLifeBalance', hue='Attrition',palette='cubehelix')
plt.title("Attrition by Work-Life Balance")
plt.show()
```

```python
sns.histplot(data=df,x='YearsAtCompany',bins=20, kde=True, hue='Attrition')
plt.title("Years at Company vs Attrition")
```
